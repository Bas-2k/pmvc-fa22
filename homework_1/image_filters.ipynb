```python
from __future__ import annotations
from collections.abc import Callable, Collection, Iterable
import itertools
import pickle

import numpy as np
from scipy.signal import convolve2d
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from skimage.color import rgb2gray

%matplotlib inline


def plot_image(image: np.ndarray, **kwargs) -> mpl.figure.Figure:
    """Utility function to plot an image.

    Args:
        image: a 2-D numpy array
    """
    fig = plt.figure()
    plt.axis('off')
    plt.imshow(image, cmap=plt.gray(), **kwargs)
    return fig


def surf(X, Y, Z, **kwargs):
    """Plot 3D data as surface, similar to surf(X,Y,Z) of http://www.mathworks.com/help/matlab/ref/surf.html"""
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, **kwargs)
    return ax
```

## Utilities

### Create sinusoid stimuli

A sinusoid stimulus $I(\mathbf{x})$ is described by

$I(\mathbf{x}) = A \cos(\mathbf{\omega} \mathbf{x} + \rho) $

Here,

- $A$ is the amplitude
- $\mathbf{\omega}$ is the frequency
- $\rho$ is the phase


```python
def create_sinusoid(
    size: tuple[int, int] = (32, 32),
    amplitude: float = 1,
    frequency: tuple[float, float] = (np.cos(np.pi / 4), np.sin(np.pi / 4)),
    phase: float = np.pi / 2
):
    """Creates a 2D sinusoid.

    Args:
        size: size of the generated stimulus, defaults to (32, 32)
        amplitude: amplitude of the sinusoid, defaults to 1
        frequency: spatial frequencies in the horizontal and vertical directions, defaults to (np.cos(np.pi / 4), np.sin(np.pi / 4))
        phase: phase, defaults to np.pi/2

    Returns:
        2D sinusoid
    """
    width, height = size
    radius = (width // 2, height // 2)
    [x, y] = np.meshgrid(range(-radius[0], radius[0] + 1), range(-radius[1], radius[1] + 1))

    return amplitude * np.cos(frequency[0] * x + frequency[1] * y + phase)


def plot_sinusoid():
    stimulus = create_sinusoid()
    fig = plot_image(stimulus)
    fig.suptitle("sinusoid stimulus")


plot_sinusoid()
```

### Create Gabor filters

A [general Gabor filter](https://en.wikipedia.org/wiki/Gabor_filter) can be defined as

$$ g(x,y;\lambda,\theta,\psi,\sigma,\gamma) = \exp\left(-\frac{x'^2+\gamma^2y'^2}{2\sigma^2}\right)\exp\left(i\left(2\pi\frac{x'}{\lambda}+\psi\right)\right) $$

Here, we implement a Gabor filter that satisfies the neurophysiological constraints for simple cells:
   
$$ \psi (x; \omega, \theta, K) = \left[\frac{\omega^2}{ 4 \pi K^2} \exp  \{-(\omega^2/8K^2)[4(x\cdot(cos\theta, sin\theta))^2 + (x \cdot ( -sin \theta, cos \theta))^2]\} \right] \times \left[ \exp \{ iwx \cdot (cos\theta, sin\theta) \} exp(K^2/2) \right] $$

```python
def create_gabor(
    size: tuple[int, int] = (256, 256),
    omega: float = 0.3,
    theta: float = np.pi / 4,
    k: float = np.pi,
    gabor_type: str = "cosine",
):
    """Creates a 2D Gabor filter that obeys V1 simple cell constraints.

    Args:
        size: size of the generated stimulus, defaults to (256, 256)
        omega: defaults to 0.3
        theta: defaults to np.pi/4
        k: defaults to np.pi
        gabor_type: whether to generate a sine Gabor or a cosine Gabor, defaults to "cosine"

    Returns:
        2D Gabor
    """
    width, height = size
    radius = (width // 2, height // 2)
    [x, y] = np.meshgrid(range(-radius[0], radius[0] + 1), range(-radius[1], radius[1] + 1))

    x1 = x * np.cos(theta) + y * np.sin(theta)
    y1 = -x * np.sin(theta) + y * np.cos(theta)

    gaussian = (omega**2 / (4 * np.pi * k**2)) * np.exp(- omega**2 / (8 * k**2) * ( 4 * x1**2 + y1**2))

    if gabor_type == "cosine":
        sinusoid = np.cos(omega * x1) * np.exp(k**2 / 2)
    elif gabor_type == "sine":
        sinusoid = np.sin(omega * x1) * np.exp(k**2 / 2)

    gabor = gaussian * sinusoid
    return gabor


def plot_gabor_filters():
    """Plot sine and cosine Gabor filters"""
    fig, axes = plt.subplots(nrows=1, ncols=2)
    for gabor_type, ax in zip(("sine", "cosine"), axes.flat):
        gabor = create_gabor(gabor_type=gabor_type)
        ax.axis("off")
        ax.imshow(gabor, cmap=plt.gray())
        ax.set_title(f"{gabor_type} Gabor filter")


plot_gabor_filters()
```

### Create Gabor filter banks

We create a set of Gabor filters that vary in

- the type of Gabor (sine or cosine)
- $\omega$
- $\theta$

```python
def create_gabor_filter_banks(
    omegas: Iterable[float] = np.arange(0.02, 0.42, 0.1),
    thetas: Iterable[float] = np.arange(0, np.pi, np.pi / 4)
) -> dict[str, list[np.ndarray]]:
    """Create Gabor filter banks of both sine and cosine Gabors at various values of omega and theta.

    Args:
        omegas: possible values of omega, defaults to np.arange(0.02, 0.42, 0.1)
        thetas: possible values of theta, defaults to np.arange(0, np.pi, np.pi / 4)

    Returns:
        Gabor filter banks
    """
    gabor_types = ("cosine", "sine")

    filter_banks = {gabor_type: [] for gabor_type in gabor_types}

    for omega in omegas:
        for theta in thetas:
            for gabor_type in gabor_types:
                filter_banks[gabor_type].append(
                    create_gabor(
                        size=(128, 128),
                        omega=omega,
                        theta=theta,
                        gabor_type=gabor_type
                    )
                )
    return filter_banks


def plot_filter_bank(
    filter_bank: Collection[np.ndarray],
    omegas: Iterable[float] = np.arange(0.02, 0.42, 0.1),
    thetas: Iterable[float] = np.arange(0, np.pi, np.pi / 4)
) -> mpl.figure.Figure:
    """Plot all 16 Gabor filters in each filter bank."""
    fig, axes = plt.subplots(nrows=len(thetas), ncols=len(omegas), figsize=(10, 10))
    for filter_bank, ax, (omega, theta) in zip(filter_banks["sine"], axes.flat, itertools.product(omegas, thetas)):
        ax.imshow(filter_bank)
        ax.axis("off")
        ax.set_title(f"$\omega$={omega:.2f}, $\\theta$={theta:.2f}")
    fig.tight_layout()
    return fig


filter_banks = create_gabor_filter_banks()

for gabor_type, filter_bank in filter_banks.items():
    plot_filter_bank(filter_bank=filter_bank)
```

### Demo: Gaussian, Laplacian of Gaussian

```python
def plot_gaussian_and_derivatives():
    sigma = 20
    [X, Y] = np.meshgrid(np.arange(-100, 101), np.arange(-100, 101))
    Z = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(X**2 + Y**2)/(2 * sigma**2))

    dx = np.roll(Z, 1, axis=1) - Z
    dx2 = np.roll(dx, 1, axis=1) - dx

    dy = np.roll(Z, 1, axis=0) - Z
    dy2 = np.roll(dy, 1, axis=0) - dy

    LoG = -(dx2 + dy2)

    ax = surf(X, Y, Z, alpha=0.3)
    ax.set_title('Gaussian')

    ax = surf(X, Y, dx + dy, alpha=0.3)
    ax.set_title('First order derivative')

    ax = surf(X, Y, LoG, alpha=0.3)
    ax.set_title('Second order derivative (Laplacian of Gaussian)')


plot_gaussian_and_derivatives()
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load the unknown Gabor filter
unknown_gabor = pickle.load(open('/content/unknown_gabor.data', 'rb'), encoding='latin1')

def create_sinusoid(size, amplitude, frequency, phase):
    x, y = np.meshgrid(np.arange(size[1]), np.arange(size[0]))
    return amplitude * np.cos(2 * np.pi * (frequency[0] * x + frequency[1] * y) + phase)

def apply_sinusoid_to_image(image, rho, omega, theta):
    # Create a sinusoidal pattern with the given parameters
    sinusoid = create_sinusoid(
        size=image.shape,
        amplitude=1,
        frequency=(omega * np.cos(theta), omega * np.sin(theta)),
        phase=rho,
    )

    # Calculate the response of the neuron to the sinusoidal stimulus
    response = (sinusoid * image).sum()

    return response

# Experiment to find the preferred orientation, frequency, and phase
def find_gabor_parameters(image):
    best_response = -np.inf
    best_parameters = None
    orientation_responses = []  # Store responses for different orientations
    frequency_responses = []  # Store responses for different frequencies
    phase_responses = []  # Store responses for different phases

    for theta in np.arange(0, np.pi, 0.1):  # Iterate through different orientations
        orientation_response = 0
        for omega in np.arange(0.05, 1.0, 0.05):  # Iterate through different frequencies
            for rho in np.arange(0, 2 * np.pi, 0.1):  # Iterate through different phases
                response = apply_sinusoid_to_image(image, rho, omega, theta)
                orientation_response += response
                frequency_responses.append((omega, response))
                phase_responses.append((rho, response))

        orientation_responses.append((theta, orientation_response))

        if orientation_response > best_response:
            best_response = orientation_response
            best_parameters = (theta, omega, rho)

    return best_parameters, orientation_responses, frequency_responses, phase_responses

# Find the preferred orientation, frequency, and phase of the unknown Gabor filter
preferred_parameters, orientation_responses, frequency_responses, phase_responses = find_gabor_parameters(unknown_gabor)
preferred_orientation, preferred_frequency, preferred_phase = preferred_parameters

print(f"Preferred Orientation: {preferred_orientation}")
print(f"Preferred Frequency: {preferred_frequency}")
print(f"Preferred Phase: {preferred_phase}")

# Extract tuning curves for orientation, frequency, and phase
orientation_tuning_curve = [response for _, response in orientation_responses]
frequency_tuning_curve = [response for _, response in frequency_responses]
phase_tuning_curve = [response for _, response in phase_responses]

# Plot the images of the unknown Gabor filter, preferred sinusoid, and Gabor filter response
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
axes[0].imshow(unknown_gabor, cmap='gray')
axes[0].set_title("Unknown Gabor Filter")

preferred_sinusoid = create_sinusoid(unknown_gabor.shape, 1, (preferred_frequency * np.cos(preferred_orientation), preferred_frequency * np.sin(preferred_orientation)), preferred_phase)
axes[1].imshow(preferred_sinusoid, cmap='gray')
axes[1].set_title("Preferred Sinusoid")

gabor_filter_response = preferred_sinusoid * unknown_gabor
axes[2].imshow(gabor_filter_response, cmap='gray')
axes[2].set_title("Gabor Filter Response")

plt.savefig('gabor_images.png')  # Save the image with the three images

# Plot the tuning curves for orientation, frequency, and phase
plt.figure(figsize=(16, 5))
plt.subplot(131)
orientations = np.arange(0, np.pi, 0.1)
plt.plot(orientations, orientation_tuning_curve)
plt.xlabel('Orientation (radians)')
plt.ylabel('Response')
plt.title('Orientation Tuning Curve')

plt.subplot(132)
frequencies = np.arange(0.05, 1.0, 0.05)
frequency_responses_mean = np.mean(np.array(frequency_tuning_curve).reshape(-1, len(frequencies)), axis=0)
plt.plot(frequencies, frequency_responses_mean)
plt.xlabel('Frequency')
plt.ylabel('Response')
plt.title('Frequency Tuning Curve')

plt.subplot(133)
phases = np.arange(0, 2 * np.pi, 0.1)
phase_responses_mean = np.mean(np.array(phase_tuning_curve).reshape(-1, len(phases)), axis=0)
plt.plot(phases, phase_responses_mean)
plt.xlabel('Phase (radians)')
plt.ylabel('Response')
plt.title('Phase Tuning Curve')

plt.tight_layout()
plt.savefig('tuning_curves.png')  # Save the tuning curve image

plt.show()

```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load the unknown LoG data from a file
unknown_LoG = pickle.load(open("/content/unknown_LoG.data", "rb"), encoding="latin1")

# Define a range of sigma values to test
sigma_values = np.linspace(0.1, 10, 100)  # Adjust the range and step as needed
responses = []

# Apply Laplacian of Gaussian with different sigma values to sinusoids
for sigma in sigma_values:
    # Create a synthetic sinusoidal pattern (adjust as needed)
    frequency = 0.1  # Adjust the frequency as needed
    amplitude = 1.0  # Adjust the amplitude as needed
    x = np.linspace(-100, 100, unknown_LoG.shape[1])
    y = np.linspace(-100, 100, unknown_LoG.shape[0])
    X, Y = np.meshgrid(x, y)
    sinusoid = amplitude * np.sin(2 * np.pi * frequency * X)

    # Apply the Laplacian of Gaussian filter
    log_filtered_sinusoid = cv2.GaussianBlur(sinusoid, (0, 0), sigmaX=sigma)
    log_filtered_sinusoid = cv2.Laplacian(log_filtered_sinusoid, cv2.CV_64F)

    # Measure the response (variance)
    response = np.var(log_filtered_sinusoid)
    responses.append(response)

# Find the sigma that maximizes the response (variance)
best_sigma = sigma_values[np.argmax(responses)]

# Plot the responses for different sigma values
plt.plot(sigma_values, responses)
plt.xlabel('Sigma')
plt.ylabel('Response (Variance)')
plt.title('Response to Laplacian of Gaussian Filter')
plt.show()

print(f"The optimal sigma is: {best_sigma}")
```
